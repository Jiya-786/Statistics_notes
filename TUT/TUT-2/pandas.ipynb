```python
#creating dataframe
data = {
    'Student_ID': [101, 102, 103, 104, 105, 106, 107, 108],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah'],
    'Age': [20, 21, 19, 22, 20, 23, 19, 21],
    'Gender': ['F', 'M', 'M', 'M', 'F', 'M', 'F', 'F'],
    'Midterm_Score': [85, 78, 92, 60, 88, np.nan, 95, 82],
    'Final_Score': [90, 80, 85, 65, 92, 70, 98, 85],
    'Attendance_Rate': [0.95, 0.80, 0.90, 0.60, 0.95, 0.50, 1.00, 0.85]
}

df=pd.DataFrame(data)
df
```

```python
#inspecting data
print("First 5 rows: \n")
display(df.head())               # it is by default 5. it prints first 5 rows

print("Dataframe info: \n")
display(df.info())               # writing display is optional

print("Dataframe summary: \n")
df.describe()

print("Dataframe shape: \n")
df.shape
```

```python
#indexing and selection
#two primary ways .loc() -> label-based -> select by column name or row label
# and .iloc() -> integer-based -> select by index position (0 to length-1)

ages=df['Age']
print("---Age column---")
print(ages.head(3))

scores=df[['Name','Midterm_Score','Final_Score']]
print("\n---Name and Scores---")
display(scores.head(3))

print("\n---using .loc---")
display(df.loc[0:2,['Name','Attendance_Rate']])

print("\n---using .iloc()---")
display(df.iloc[0:3,0:3])
```

```python
older_students=df[df['Age']>20]
print("---Older Students---")
display(older_students)

high_scoring_females=df[(df['Gender']=='F') & (df['Final_Score']>85)]
print("---Highi scoring female students---")
display(high_scoring_females)
```

```python
# data cleaning
print("Missing values per column:")
print(df.isnull().sum())

df_dropped=df.dropna()
print(f"\nShape after dropping NaNs: {df_dropped.shape}")

midterm_mean=df['Midterm_Score'].mean()
df['Midterm_Score']=df['Midterm_Score'].fillna(midterm_mean)

print("\nData after filling NaNs: ")
display(df)
```

```python
#creating new columns
df['Total_Score']=df['Midterm_Score']+df['Final_Score']

df['Passed']=np.where(df['Total_Score'] >= 150, 'Yes', 'No')
display(df[['Name','Total_Score','Passed']])
```

```python
# grouping and aggregation

# This is critical for statistical analysis. We use .groupby() to split the data into groups
# and apply a function (mean, sum, count) to each group.
# We can also check the Correlation Matrix to see relationships between variables.

gender_group = df.groupby('Gender')[['Midterm_Score', 'Final_Score', 'Attendance_Rate']].mean()
print("--- Average Scores by Gender ---")
display(gender_group)

print("\n--- Pass/Fail Count ---")
print(df['Passed'].value_counts())

print("\n--- Correlation Matrix ---")
correlation = df[['Age', 'Midterm_Score', 'Final_Score', 'Attendance_Rate']].corr()
display(correlation)
```

```python
# saving the data

df.to_csv("processed_student_data_csv",index=False)
print("File successfully saved")


```

```python
# Exercise 1: Filtering and Sorting

# Using the df (Student Data) we created earlier:

# Create a new DataFrame called consistent_students
# containing only students with an Attendance_Rate greater than 0.85.

# Sort this new DataFrame by Final_Score in descending order (highest score first).
# Display the result.

consistent_students=df[df['Attendance_Rate']>0.85]
df_sorted=consistent_students.sort_values(by='Final_Score',ascending=False)
display(df_sorted)
```

```python
# Exercise 2:Cleaning "Corrupt" Data**
# In real-world statistics, sensors or systems often record errors as
# specific numbers (like -999 or -1) instead of leaving them empty.

# Scenario: You have temperature data from a sensor.
# However, the sensor malfunctioned a few times and recorded -999.

# Your Tasks:
# Create the DataFrame using the provided dictionary.
# Replace the value -999 with np.nan (Look at the Appendix Section C).
# Fill the missing values (NaN) with the median of the Temperature column (not the mean!).
# Display the cleaned DataFrame.*

sensor_data = {
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'],
    'Temperature': [22.5, 24.1, -999, 23.0, -999, 25.5, 26.0]
}
df=pd.DataFrame(sensor_data)
df=df.replace(-999,np.nan)         # only df.replace(-999,np.nan) will not work.
                                   # df will remain unchanged
                                   # else we will have to use df.replace(-999,np.nan,inplace=True)
temp_median=df['Temperature'].median()
df['Temperature']=df['Temperature'].fillna(temp_median)
display(df)

```

```python
# Exercise 3: Grouping and Variability
# In statistics, the mean isn't always enough.
# We often need to know the spread (Standard Deviation) or the middle value (Median)
# to understand the data better.

# Scenario: You have data on employee salaries across different departments.

# Your Tasks:
# Create the DataFrame.
# Group the data by Department.
# Extract the Salary column from the groups.
# Calculate two separate statistics for each department:
# The Median Salary.
# The Standard Deviation (std) of the Salary (Look at Appendix Section B).
# Print both results.

employee_data = {
    'Department': ['Sales', 'Tech', 'Sales', 'Tech', 'HR', 'Sales', 'Tech', 'HR'],
    'Employee': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Salary': [50000, 80000, 52000, 85000, 60000, 51000, 82000, 59000]
}

df=pd.DataFrame(employee_data)
grouped_data=df.groupby('Department')
salary=grouped_data['Salary']
median_salary=salary.median()
std_dev_salary=salary.std()
print(f"Median Salary:\n {median_salary}")
print(f"Std Dev of salary:\n {std_dev_salary}")

```
