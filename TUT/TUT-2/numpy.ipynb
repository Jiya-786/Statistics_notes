```python
import numpy as np
```

```python
#creating arrays in numpy
py_list=[1,2,3,4,5]
arr=np.array(py_list)

print("Python list: ",py_list)
print("Numpy array: ",arr)
print("Shape of array: ",arr.shape)

#creating a 2D array
matrix=np.array([[1,2,3],[4,5,6]])
print("2D matrix\n: ",matrix);
print("Matrix shape: ",matrix.shape)

```

```python
#slicing and indexing in numpy
matrix=np.array([10,20,30,40])
print(matrix[0])
print(matrix[-1])
print(matrix[1:4])
```

```python
#slicing and indexing in 2D array
matrix=np.array([[1,2,3],[4,5,6]])
print(matrix[0,1]) #row 0 column 1
print(matrix[:,1]) #all rows column 1
```

```python
#data type in numpy
matrix1=np.array([1,2,3])
print(matrix1.dtype)

matrix2=np.array([1,2,3],dtype=float)
print(matrix2.dtype)
```

```python
#shape vs size vs ndim
""" shape -> structure
    size -> tot no.of elts
    dim -> no.of dimensions """

matrix=np.arange(12)
print("og matrix: ",matrix)
matrix=matrix.reshape(3,4)
print("rearranged matrix: \n",matrix)
print("Shape: ",matrix.shape)
print("Size: ",matrix.size)
print("Dimension: ",matrix.ndim)   # here we have row,column hence dimension is 2.
                                   # going further we can also find another 3rd dimension
                                   # other than row and column although the matrix can be hard to imagine

```

```python
# 3x3x3 matrix ie 3 dimensions matrix, one extra dimension other than row and column
matrix=np.arange(1,28).reshape(3,3,3)
print(matrix)
print("Required elt: ",matrix[0,2,1])    #indexing of rows and columns also starts from 0
```

```python
# 4x4x4 matrix
matrix=np.arange(1,82).reshape(3,3,3,3)
print(matrix)

print("Required elt: ",matrix[0,2,2,1])     #open from outside bracket to inside bracket
```
