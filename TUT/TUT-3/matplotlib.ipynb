```python
#matplotlib is a python library for drawing graphs,charts and plots
import matplotlib.pyplot as plt
import numpy as np

# line plot-

days = np.array([1, 2, 3, 4, 5, 6, 7])
temperature = np.array([22, 24, 25, 23, 26, 28, 27])

plt.figure(figsize=(8,4))
plt.plot(days,temperature)

plt.title("Weekly temperature trend")
plt.xlabel("days")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
# Box-plot
# Suppose I give you 200 numbers (marks of 200 students). You cannot look at all 200 numbers.

# So we summarize the data using 5 numbers: this is called the Five Number Summary
data_group_A = np.random.normal(100, 10, 200)    #Group A: 200 values around 100
data_group_B = np.random.normal(80, 30, 200)     #200 values around 80 but more spread out

data_to_plot = [data_group_A, data_group_B]
plt.boxplot(data_to_plot, labels=['Group A', 'Group B'])
plt.title("Comparison of Distributions")
plt.ylabel("Values")
plt.grid(axis='y')
# plt.grid()
# plt.grid(axis='x')

plt.show()
```

```python
# simplest way of visualizing gradient descent

# In statistics and machine learning, we often try to find the "best" parameters for a model.
# "Best" usually means the parameters that minimize the error (or "Cost Function").

# Gradient Descent is an optimization algorithm used to find this minimum value.
# Think of it as standing on top of a hilly terrain (the cost surface) and taking steps downhill in the steepest direction until you reach the lowest valley (minimum error).


# creating the bowl surface
x=np.linspace(-5,5,50)
y=np.linspace(-5,5,50)
X,Y=np.meshgrid(x,y)
Z=X**2+Y**2

# starting point
cx,cy=-4.0,4.0
lr=0.1

# store path
hx, hy, hz=[],[],[]

# do gradient descent (walk downhill)
for _ in range(20):
  z=cx**2+cy**2
  hx.append(cx)
  hy.append(cy)
  hz.append(z)

  # slope
  gx=2*cx
  gy=2*cy

  # step down
  cx=cx- lr*gx
  cy=cy-lr*gy


# plot
ax=plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,alpha=0.5)
ax.plot(hx,hy,hz,marker='x')             # path

plt.show()


```
